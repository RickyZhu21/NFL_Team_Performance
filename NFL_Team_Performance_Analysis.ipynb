## Setup

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import patsy
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

Function to standardize the team name for both datasets:
- Leaving only team names without the city name beside them.
- Rename some team names that are not representational for the dataset(Ex: team name columns that exist some only city names)
- Remove duplicate team name that is represented in two different forms. (Combine into single team name)

```python
def rename_team(name):
    teams = ['Arizona Cardinals','Atlanta Falcons','Baltimore Ravens','Buffalo Bills','Carolina Panthers',
             'Chicago Bears','Cincinnati Bengals','Cleveland Browns','Dallas Cowboys','Denver Broncos',
             'Detroit Lions','Green Bay Packers','Houston Texans','Indianapolis Colts','Jacksonville Jaguars',
             'Kansas City Chiefs','Las Vegas Raiders','Los Angeles Chargers','Los Angeles Rams','Miami Dolphins',
             'Minnesota Vikings','New England Patriots','New Orleans Saints','New York Giants','New York Jets',
             'Philadelphia Eagles','Pittsburgh Steelers','San Francisco 49ers','Seattle Seahawks','Tampa Bay Buccaneers',
             'Tennessee Titans','Washington Commanders']
    name = name.lower()
    for word in name.split():
        for team in teams:
            team = team.lower()
            #some teams were renamed
            if word == 'redskins':
                return 'commanders'
            #some teams moved cities
            if word == 'oakland':
                return 'raiders'
            if word == 'diego':
                return 'chargers'
            if 'louis' in word:
                return 'rams'
            #some teams have overlapping words
            if 'green' == word:
                return 'packers'
            if word in team:
                name = team.split()[-1]
    return name
```

## Dataset: National Football League League Rankings 2000-2021

```python
nfl_stats = pd.read_csv('standings.csv')
nfl_stats.head()
```

## Data Cleaning: National Football League League Rankings 2000-2021

Drop irrelevant columns for the Dataset

```python
columns_to_drop = ['Unnamed: 0', 'PCT', 'Road', 'Pct', 'Pct.1', 'Last 5']
nfl_stats.drop(columns_to_drop, axis=1, inplace=True)
nfl_stats.head()
```

Rename some columns for the Dataset to avoid confusions when looking at the data

```python
nfl_stats = nfl_stats.rename(columns={'W': 'Win', 'L': 'Loss', 'T': 'Tie', 'PF': 'Team Total Point', 
                                      'PA': 'Opponent Total Point', 'Net Pts': 'Net Points', 'Strk': 'Streak', 'NFL Team': "Team"})
nfl_stats.head()
```

The original Dataset contained variables that had multiple categories within that variable, therefore to avoid confusion about what each value represented, we separated these embedded variables into their own columns. 
- For instance, when playing at home, each team’s wins, loses, and ties, were represented within the same variable under “Home”. We separated these variables into home_win, home_loss, and home_tie.

```python
columns_to_process = ['Home', 'Div', 'Conf', 'Non-Conf']

for col in columns_to_process:
    # Splitting the string into three columns
    new_cols = nfl_stats[col].str.split(' - ', expand=True)
    
    # Renaming the columns
    new_cols.columns = [f"{col}_{part}" for part in ['win', 'loss', 'tie']]
    
    # Concatenating new columns to the original DataFrame and dropping the old column
    nfl_stats = pd.concat([nfl_stats, new_cols], axis=1).drop(columns=[col])
```

```python
nfl_stats.head()
```

The “streaks” variable contained some strings such as “L” for loss or “W” for win in conjunction with integers representing the number of wins or losses. We decided to change the data to strictly integers, using negative signs to represent losses.

```python
def convert_streak(streak):
    if 'L' in streak:
        return -int(streak.replace('L', '')) #Convert L from string to int with negatives
    elif 'W' in streak:
        return int(streak.replace('W', '')) #Convert W from string to int. 
    else:
        return 0  # Default case if no W or L

nfl_stats['Streak'] = nfl_stats['Streak'].apply(convert_streak)
nfl_stats.head()
```

Lastly, we standardize the team name for the Dataset using the function in the Setup section to keep everything standard for team names across both datasets. 

```python
nfl_stats['Team'] = nfl_stats['Team'].apply(rename_team)
nfl_stats.head()
```

## Data Cleaning: Super Bowl History 1967-2021

```python
sb = pd.read_csv('superbowl.csv')

#convert date to just year
def get_year(date):
    return date.split()[-1]
sb['Date'] = sb['Date'].apply(get_year)
sb = sb.rename(columns = {'Date' : 'Year'})

#remove unneccessary columns
sb = sb[['Year', 'Winner', 'Loser']]

#add data from 2021
sb.loc[len(sb.index)] = ['2021', 'Tampa Bay Buccaneers', 'Kansas City Chiefs']

#rename teams
sb['Winner'] = sb['Winner'].apply(rename_team)
sb['Loser'] = sb['Loser'].apply(rename_team)

#convert years to int
sb['Year'] = sb['Year'].astype('int')

#Total 52 rows(1967-2021), for clarity using head()
sb.head()
```

## Merging

First, we are merging the super bowl stats as a column to the dataset: National Football League League Rankings 2000-2021 to demonstrate whether each team has went to the super bowl or not.

```python
#Function: 1 represents that NFL team has went to the super bowl in the respective season. 0 otherwise. 
def in_sb(team, year):
    if team in sb[sb['Year']==year].values:
        return 1
    else:
        return 0
    
#add superbowl row to test2
nfl_stats['super_bowl'] = nfl_stats.apply(lambda row : in_sb(row['Team'], row['Year']), axis = 1)
nfl_stats.head()
```

### Inferential Analysis

#### Linear regression

We decided to generate a linear regression plot to check the linear relationship between the team’s financial power such as Total expenses, to the season performance of the NFL teams in terms of their average ranks. The results of the linear regression analysis are performed using OLS regression analysis.


```python
#Linear regression report between the teams' average rank and their average total expenses over past 20 seasons
mod = smf.ols(formula='Rank ~ Total_Expenses', data=average_stats)
res = mod.fit()
print(res.summary())
```

The negative linear relationship demonstrated by this analysis means that for every one unit increase in the total expense, the average rank for the team will decrease by 0.0384 explicitly in terms of numerical values. **Please Note: Rank 1 is the best rank in the NFL for all 32 teams in the NFL per year, and Rank 32 is the worst rank for the team in the NFL for all 32 teams in the NFL per year.** The decrease in numbers for rank explicitly means that as a team gets better, in linear terms, their rank number will eventually go lower and lower up until Rank 3, 2, 1, as Rank 1 is the best team in terms of performance in the NFL season. 

Based on the p-values returned by the linear regression comparing the teams’ seasonal performance and their financial power as total expenses for the past 20 seasons, the result of this linear regression analysis is not statistically significant (p-value > 0.05), meaning that the likelihood of a linear relationship between the team’s total expense and their average rank is unlikely. This seems consistent with our average data across 20 years as the average ranks of each team were all similar with Browns having the lowest overall rank by a very large margin. The expenses for each team were more spread out than the average rank of the teams.

# Discussion
After conducting our analysis using the datasets we collected, we can conclude that there is not a statistically significant relationship between the season performance of an NFL team and the team’s financial power over the past 20 seasons. In order to measure team performance, we analyzed each team’s performance with their average ranking and the proportion of occurrences that each team has made over the 20 seasons. A team’s financial power was measured through the total expense, which is the summation of all of the expenses that goes into running the teams, and player expenses, which measures the amount of money each team has allocated to sign different NFL athletes. 

We had hypothesized that there would be a positive correlation between a team’s financial income and their average performance on the field. We believed that a larger fund would result in access to the best athletes, coaches, facilities, and trainers which in turn would reflect in the team’s overall performance. Throughout this study we utilized descriptive statistics, average calculations over the season, histograms, bar plots for visual representation of the data, and inferential statistics, specifically an Ordinary Least Squares (OLS) linear regression analysis, to investigate the relationship between team expenses and performance.Data was gathered from three primary sources: NFL Official Statistics, Rodney Fort's Sports Business Data Pages, and Super Bowl History. These datasets were cleaned, merged, and preprocessed to ensure consistency and relevance to the research question.

However, our analysis, particularly the linear regression, indicated a weak negative correlation between total team expenses and their average rank, with an R-squared value of 0.033. As total expense increases, rank decreases(Rank 1 being best, Rank 32 being worst). This suggests that increased financial expenditure is marginally associated with better average team rank. However, this relationship was not statistically significant (p-value > 0.05). This suggests that while there is some relationship between finances and performance, it is not strong enough to assert that increased financial investment will reliably improve a team's performance. We found that some teams, notably the Cowboys, had high total expenses yet found themselves ranking around the midrange and have not made any appearance at the superbowl within the last 5 years. Meanwhile some teams with relative financial power to other teams (around the 2-3 hundred million dollar total expenses) are able to significantly outrank their opponents. For instance, the Patriots who have an average total expense of 261 million dollars, which is comparable to the majority of the NFL teams, have been able to rank an average of 6 in end of season rankings and attend the Super Bowl nine times within a twenty year period. In our Super Bowl attendance bar plot, there does appear to be a small relationship between a team’s average total expenses and their appearance at the Super Bowl since the team that has appeared the most times seems to have a slightly higher total expense than the teams that have appeared less frequently or not at all. Despite this, our statistical analysis does not support the conclusion that an NFL team’s financial power corresponds to their on field performance. 

No study goes without its own set of limitations, and ours is no expectation. While we would have liked to look at the course and performance of different teams throughout all of NFL history, we were only able to gather NFL financial data starting in the 2000s. This limited our study to a range of 20 years as opposed to starting from the year the NFL was founded in 1920. Furthermore, we are reliant on the financial transparency and accuracy of the NFL to create these conclusions. Therefore, if the numbers we are working with are not a true representation of the team’s financial information, our analysis may be skewed. Lastly, in terms of performance, this study does not account for external factors such as team culture, management decisions, or the impact of individual players and injuries, which can all significantly influence performance.
